```python
import matplotlib.pyplot as plt
all_tasks_fr = [
    "Construction des donnes", "Fonction de connexion", "Fonction d'inscription",
    "Fonction de rinitialisation du mot de passe", "FAQ (Partie complte)",
    "FAQ (Ajout de nouvelles questions)", "Playlist (Frontend complt)",
    "Dveloppement backend (Playlist)", "Cration de playlist", "Ajout de runions",
    "Rsolution des problmes avec le professeur et $_SESSION",
    "Page de localisation", "Page Qui Sommes Nous", "Page Capteur", "Page Information",
    "Gestion du niveau d'accs", "Amlioration de la convivialit", "Optimisation artistique"
]

all_weights = [3, 4, 4, 4, 3, 1, 4, 2, 1, 4, 0.25, 0.25, 0.5, 2, 0.5, 2, 1, 1]


all_percentages = [
    100, 95, 95, 90, 90, 0, 90, 0, 0, 90, 0, 0, 0, 0, 0, 0, 0, 0
]


weighted_total_progress_updated = sum(p * w for p, w in zip(all_percentages, all_weights)) / sum(all_weights)


fig, ax = plt.subplots(figsize=(14, 10))
ax.barh(all_tasks_fr, all_percentages, color='teal')
ax.set_xlabel('Pourcentage de compltion (%)')
ax.set_title('Graphique de progression du projet global aprs ajustements')
ax.set_xlim(0, 100)


ax.axvline(x=weighted_total_progress_updated, color='red', linestyle='--', label=f'Progression globale moyenne: {weighted_total_progress_updated:.2f}%')
ax.legend()


for i, (task, pct) in enumerate(zip(all_tasks_fr, all_percentages)):
    ax.text(pct + 1, i, f"{pct}% (Poids: {all_weights[i]})", color='darkblue', va='center')


plt.show()


weighted_total_progress_updated
```
